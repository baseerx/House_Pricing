```python
impacting_columns=corr[corr['price_label']>=0.3]
```

```python
selected_columns_ml=impacting_columns.index.to_list()
```

```python
selected_columns_ml
```

```python
selected_columns_ml.remove('price_label')
```

```python
print(selected_columns_ml)
```

```python
df_selected=df[selected_columns_ml+['price_label']]
```

```python
df_selected.head()
```

```python
x=df_selected.drop('price_label',axis=1)
```

```python
x
```

```python
y=df_selected['price_label']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
```

```python
x_train_scaled=StandardScaler().fit_transform(x_train)
x_test_scaled=StandardScaler().fit_transform(x_test)
```

```python
model=LinearRegression()
model.fit(x_train_scaled,y_train)
pred=model.predict(x_test_scaled)
```

```python
score=mean_squared_error(y_test,pred)
```

```python
score
```

```python
rfr = RandomForestRegressor(n_estimators=300, max_depth=20, random_state=42)
rfr.fit(x_train, y_train)
```

```python
newscore=rfr.predict(x_test)
```

```python
score=mean_squared_error(y_test,newscore)
```

```python
score
```

```python
r2score=r2_score(y_test,pred)
```

```python
print(r2score)
```

```python

# Combine into a DataFrame
df_plot = pd.DataFrame({'Actual': y_test, 'Predicted': pred})

fig = px.scatter(df_plot, x='Actual', y='Predicted',
                 title='Actual vs Predicted Prices',
                 labels={'Actual':'Actual Price', 'Predicted':'Predicted Price'})
fig.add_shape(type="line", x0=df_plot['Actual'].min(), y0=df_plot['Actual'].min(),
              x1=df_plot['Actual'].max(), y1=df_plot['Actual'].max(),
              line=dict(color="red", dash="dash"))  # perfect prediction line
fig.show()
```
